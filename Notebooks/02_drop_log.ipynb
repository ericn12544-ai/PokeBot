```python
import pandas as pd
from datetime import datetime
import os

# Load product master (read-only)
products = pd.read_csv("../data/products.csv")

# show the DataFrame head (avoid using backticks)
print(products.head())
```

```python
# Drop event table
drop_events = pd.DataFrame(columns=[
    "drop_id",
    "retailer",
    "source",            # app / email / in_person
    "zip_code",          # ✅ NEW
    "price_observed",    # ✅ NEW
    "observed_at",       # ✅ NEW (ISO timestamp)
    "notes"
])


# Mapping table: which products were in which drop
drop_products = pd.DataFrame(columns=[
    "drop_id",
    "product_id"
])

drop_events, drop_products
```

```python
drop_id = 1
# Repeatable code to log a drop event

from datetime import datetime

drop_events.loc[len(drop_events)] = [
    drop_id,
    "Target",
    "app",
    "60491",          # zip_code
    49.99,            # price_observed
    datetime.utcnow().isoformat(),
    "ETB restock via Target app"
]
# Link the drop to a product from Product Master
drop_products.loc[len(drop_products)] = [
    drop_id,
    1  # product_id from products.csv
]

drop_events, drop_products
```

```python
os.makedirs("../data", exist_ok=True)

drop_events.to_csv("../data/drop_events.csv", index=False)
drop_products.to_csv("../data/drop_products.csv", index=False)

print("✅ drop_events.csv and drop_products.csv created")
```
