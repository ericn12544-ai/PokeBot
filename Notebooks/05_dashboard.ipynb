```python
import pandas as pd

products = pd.read_csv("../data/products.csv")
scored_drops = pd.read_csv("../data/scored_drops.csv")
store_patterns = pd.read_csv("../data/store_patterns.csv")
alert_history = (
    pd.read_csv("../data/alert_history.csv")
    if (pd.io.common.file_exists("../data/alert_history.csv"))
    else pd.DataFrame()
)

scored_drops.head()
```

```python
ACTION_THRESHOLD = 7

action_now = (
    scored_drops
    .query("drop_score >= @ACTION_THRESHOLD")
    .sort_values("drop_score", ascending=False)
    .head(10)
)

action_now[[
    "product_name",
    "retailer",
    "zip_code",
    "price_observed",
    "msrp",
    "drop_score",
    "source",
    "observed_at"
]]
```
